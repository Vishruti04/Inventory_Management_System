# Purchase and update the json file

```python
prod_id = int(input("Enter the product id: "))
ui_quant = int(input("Enter the item quantity: "))

if products[prod_id]['quantity']>0:
    products[prod_id]['quantity'] -= ui_quant
else:
    print("Product out of stock")
    
print("product", products[prod_id]['item'])
print("price", products[prod_id]['cost'])
print("Total Billing Amount", products[prod_id]['cost'] * ui_quant)

#update the json file

Products_list = json.dumps(products)
with open('inventory.json','w') as file:
    file.write(Products_list)
    file.close()
```
